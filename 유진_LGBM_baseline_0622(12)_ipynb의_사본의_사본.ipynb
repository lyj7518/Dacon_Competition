# 하이퍼파라미터 튜닝

```python
import lightgbm as lgb
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

## RandomizedSearchCV 사용

```python
# 기본 모델
pipe = make_pipeline(
    lgb.LGBMRegressor(random_state=42, max_depth=7)
)

pipe.fit(X_train, y_train)

y_pred_train = pipe.predict(X_train)
y_pred_val = pipe.predict(X_val)

print('훈련 점수: ', mse(y_train, y_pred_train)**0.5)
print('검증 점수: ', mse(y_val, y_pred_val)**0.5)
```

```python
# RandomizedSearchCV
pipe = make_pipeline(
    SimpleImputer(),
    StandardScaler(),
    lgb.LGBMRegressor(random_state=1, n_jobs=-1))

dists = {
    'simpleimputer__strategy': ['mean', 'median','most_frequent'],
    'lgbmregressor__learning_rate' : [0.05, 0.10,0.15,0.20,0.25], 
    'lgbmregressor__num_leaves' : [20, 40, 60, 80, 100],
    'lgbmregressor__min_child_samples' : [5, 10, 15, 20],
    'lgbmregressor__max_depth': [5, 10, 15, 20, None],
    'lgbmregressor__n_estimators' : [50, 100, 500, 1000],
    'lgbmregressor__reg_alpha' : [0, 0.01, 0.02, 0.03]
    }

clf = RandomizedSearchCV(
    pipe, 
    param_distributions=dists, 
    n_iter=10, 
    cv=5, 
    scoring='neg_root_mean_squared_error',  
    verbose=1,
    n_jobs=-1
)

clf.fit(X_train, y_train)
```

```python
print("최적 하이퍼파라미터 : ", clf.best_params_)
print('rmse: ', clf.best_score_)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_val = clf.predict(X_val)
print('훈련 점수: ', mse(y_train, y_pred_train)**0.5)
print('검증 점수: ', mse(y_val, y_pred_val)**0.5)
```

## GridSearchCV

```python
# GridSearchCV
pipe = make_pipeline(
    SimpleImputer(),
    StandardScaler(),
    lgb.LGBMRegressor(random_state=2, n_jobs=-1))

dists = {
    'simpleimputer__strategy': ['mean', 'median','most_frequent'],
    'lgbmregressor__learning_rate' : [0.05, 0.10,0.15,0.20,0.25], 
    'lgbmregressor__num_leaves' : [20, 40, 60, 80, 100],
    'lgbmregressor__min_child_samples' : [5, 10, 15, 20],
    'lgbmregressor__max_depth': [5, 10, 15, 20, None],
    'lgbmregressor__n_estimators' : [50, 100, 500, 1000],
    'lgbmregressor__reg_alpha' : [0, 0.01, 0.02, 0.03]
    }

clf = GridSearchCV(
    pipe, 
    param_grid=dists, 
    cv=2, 
    scoring='neg_root_mean_squared_error',  
    verbose=1,
    n_jobs=-1
)

clf.fit(X_train, y_train)
```

```python
print("최적 하이퍼파라미터 : ", clf.best_params_)
print('rmse: ', clf.best_score_)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_val = clf.predict(X_val)
print('훈련 점수: ', rmse(y_train, y_pred_train))
print('검증 점수: ', rmse(y_val, y_pred_val))
```

## Early Stopping 사용

```python
scaler = StandardScaler()
imputer = SimpleImputer(strategy='median')

X_train_imputed = imputer.fit_transform(X_train, y_train) # 학습데이터
X_val_imputed = imputer.transform(X_val) # 검증데이터

X_train_scaled = scaler.fit_transform(X_train_imputed)
X_val_scaled = scaler.transform(X_val_imputed)

model = lgb.LGBMRegressor(
    reg_alpha = 0.03,
    num_leaves = 40,
    min_child_samples = 10,
    max_depth = 20, 
    learning_rate = 0.05,
    n_estimators = 1000,
    n_jobs=-1)

eval_set = [(X_train_scaled, y_train), 
            (X_val_scaled, y_val)]

model.fit(X_train_scaled, y_train, 
          eval_set=eval_set,
          early_stopping_rounds=100,
          eval_metric="rmse",
          verbose=True
         )           
```

```python
y_pred_train = model.predict(X_train_scaled)
y_pred_val = model.predict(X_val_scaled)
print('훈련 점수: ', mse(y_train, y_pred_train)**0.5)
print('검증 점수: ', mse(y_val, y_pred_val)**0.5)
```

- cv 돌렸을 때 나온 best_prams 넣었을 때의 score가 가장 좋당
- n_estimator가 너무 많은 것 같지만 n_estimator=5000, early_stopping_rounds=500 설정했을 때 점수가 가장 좋았다,,!
